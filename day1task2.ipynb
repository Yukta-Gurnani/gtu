# 1. Write a program to visualize the total profit of all months using line plot.


```python
import matplotlib.pyplot as plt
y = df['total_profit']
x = df['month_name']
plt.plot(x,y)
plt.show()
```

#  2. Write a program to visualize total profit of all months and show line plot with the following Style 
 properties. 
➢ Line Style dotted and Line-color should be blue
➢ Show legend at the upper right location.
➢ X label name = Month Number
➢ Y label name = Item Sold units number
➢ Add a circle marker
➢ Line marker color as yellow
➢ Line width should be 5 

```python
import matplotlib.pyplot as plt
y = df['total_profit']
x = df['month_name']
plt.plot(x,y,linestyle='dotted',color='b',marker='o',markerfacecolor='yellow',linewidth=5)
plt.legend(['blue'],loc='upper right')
plt.xlabel('Month Number')
plt.ylabel('Item Sold units number')
plt.show()
```

# 22. Write a python Matplotlib program to plot y=x^2 (parabola) and add text "Parabola $Y = x^2$" in the 
 plot.


```python
import numpy as np
x = np.arange(-1,1.25,0.25)
y = x*x
plt.plot(x,y)
plt.title("Parabola Y = x^2")
plt.show()
```
